```python
import pandas as pd
import numpy as np

df = pd.read_csv('../../data/tt_postulaciones_samp.csv')

df = df.drop('per_id', axis=1) # read in unnecessary column
```

```python
df.head().T
```

### Duplicates

```python
# No apparent duplicates in this subset
def get_shape(df):
    print(df.shape)
    return df

(df
 .pipe(get_shape)
 .drop_duplicates()
 .pipe(get_shape)
 .sample()
)
```

### Understanding our data

#### Missing Values

```python
(df
 .isna()
 .groupby(df.periodo)
 .mean()
 .style
 .format(lambda fl: f'{fl * 100:.2f} %')
)
```

#### Memory

```python
# 2221.896434
(df
 .memory_usage(deep=True)
 .apply(lambda s: s / 1000000)
 .pipe(lambda df_: pd.concat([df_, df.dtypes], axis = 1))
 .rename(columns={0:'memory', 1:'dtype'})
)
```
